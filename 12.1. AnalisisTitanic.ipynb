## Variable Dummy o Ficticia.
Una variable ficticia (también conocida como variable indicadora) es una variable numérica que representa datos categóricos, como género, raza, afiliación política, etc.

Los investigadores utilizan variables ficticias para analizar ecuaciones de regresión cuando una o más variables independientes son categóricas. La clave del análisis es expresar las variables categóricas como variables ficticias.

Técnicamente, las variables ficticias son variables dicotómicas y cuantitativas; pueden tomar dos valores cuantitativos cualesquiera. En la práctica, los resultados de la regresión son más fáciles de interpretar cuando las variables ficticias toman dos valores específicos, 1 o 0. Normalmente, 1 representa la presencia de un atributo cualitativo y 0 representa la ausencia.

El número de variables ficticias necesarias para representar una variable categórica particular depende del número de valores que puede asumir la variable categórica. Para representar una variable categórica que puede asumir k valores diferentes, un investigador necesitaría definir k - 1 variables ficticias.

Por ejemplo, supongamos que estamos interesados ​​en la afiliación política, una variable categórica que podría asumir tres valores: republicano, demócrata o independiente. Podríamos representar la afiliación política con dos variables ficticias:

X1 = 1, si es republicano; X1 = 0, de lo contrario.

```python
# Variables Dummy
def creardummy(df,var_name):
    dummy_sex=pd.get_dummies(df[var_name],prefix=var_name)
    Df=df.drop(var_name,axis=1)
    Df=pd.concat([Df,dummy_sex],axis=1)
    return Df
```
