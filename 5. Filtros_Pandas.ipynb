## Operaciones con las filas de un DataFrame


1.   Añadir una fila a un DataFrame



```python
import pandas as pd
datos={ 'Producto':['Licuadora','Ventilador','Vaporizador','Extractor'],
        'Cantidad': [24,36,39,40],
         'Precio': [2400,3500,2100,1500]}
df=pd.DataFrame(datos)
df
```

```python
datos1={ 'Producto':'Refri',
        'Cantidad': 50,
         'Precio': 2400}
datos1
```

```python
serie=pd.Series(datos1)
serie
```

```python
df1=df.append(serie,ignore_index=True)
df1
```

Fusionar, unir, concatenar y compara DataFrame

Concatenar DataFrame

```python
import pandas as pd
import numpy as np
df1=pd.DataFrame(np.random.rand(4,4),index=[0,1,2,3],columns=['A','B','C','D'])
df1
```

El segundo DataFrame

```python
df2=pd.DataFrame(np.random.rand(4,4),index=[4,5,6,7],columns=['A','B','C','D'])
df2
```

```python
df3=pd.DataFrame(np.random.rand(4,4),index=[8,9,10,11],columns=['A','B','C','D'])
df3
```

```python
dfc=pd.concat([df1,df2,df3],axis=0)
```

```python
dfc
```

keys: Permite construir un indice jerárquico usando las claves pasadas como nuvel más externo.

```python
resultado=pd.concat([df1,df2,df3],keys=['X','Y','Z'])
resultado
```

```python
resultado.loc[['Y','X']]
```

```python
Dataframe=[df*100 for df in [df1,df2,df3]]
resultados=pd.concat(Dataframe)
resultados
```

Definamos un cuarto DataFrame

```python
df4=pd.DataFrame(np.random.rand(4,3),index=[2,3,6,7], columns=['B','D','F'])
df4
```

```python
resultado=pd.concat([df1,df4],axis=1)
resultado
```

Aplicando join='inner' en el parametro concat

```python
resultado=pd.concat([df1,df4], axis=1, join='inner')
resultado
```

Reindexando a partir del dataframe df1

```python
resultado=pd.concat([df1,df4], axis=1).reindex(df1.index)
resultado
```
