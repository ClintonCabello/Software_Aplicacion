## Estructura DataFrame

DataFrame define un conjuto de datos estructurados en forma de tabla, donde cada columna es un objeto de tipo serie

a) Crear un DataFrame a partir de un diccionario de listas

```python
import pandas as pd

datos={ 'nombre':['Tania','Marcia','Juan','Miguel'],
         'sexo':['femenino','femenino','masculino','masculino'],
         'edad':[23,43,12,44]}
df=pd.DataFrame(datos)
df
```

### Tratamiento de datos faltantes en un DataFrame

Ejemplo Un DataFrame con datos faltantes


```python
import pandas as pd
import numpy as np
datos={ 'nombre':['omeprazol','panadol','paracetamol',None,'lansoprazol'],
         'precio':[ 234,123,np.nan,134,np.nan],
          }
df=pd.DataFrame(datos)
df
```

```python
df.isnull().sum()
```

a) El método fillna() : sustituye en los valores faltantes por el valor que indica en el parámetro

```python
df.fillna(0)
```

b) Para sustituir valores faltantes se usa la función replace()

```python
df.replace(np.nan,1000)
```

c) Si usamos el método interpolate() sustituye el valor nan por valores interpoladas

```python
df.interpolate()
```

d) El método dropna() elimina las filas que contiene los datos faltantes

```python
df.dropna()
```
