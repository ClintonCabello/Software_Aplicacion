### Filtrado de una serie

```python
# Pacientes infectados por COVID 19 de diferentes paises
import pandas as pd
S=pd.Series({'Perú':2343,'Colombia':234,'Brasil':4324,'España':2345,'Francia':231,'Italia':234})
S
```

```python
# Paises que tienen pacientes infectados mas de 2000
S[S>2000]
```

```python
S[(S>1000) & (S<2500)]
```

```python
S[(S>1000) | (S<2500)]
```

Ordenar una serie

```python
S.sort_values(ascending=True)
```
