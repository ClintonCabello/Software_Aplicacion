## FUSIONAR UNIR, CONCAENAR Y COMPARAR DATAFRAME

* CONCATENAR DATAFRAME

  La funciòn concat() permite realizar operaciones de concatenaciòn de series o DataFrame, para se hace uso de operaciones bàsicas del algebra de conjuntos

```python
# Ejemplo: Definamos DataFrame a fin de realizar las concatenaciones
import pandas as pd

df1=pd.DataFrame({
    "A":['a0','a1','a2','a3'],
    "B":['b0','b1','b2','b3'],
    "C":['c0','c1','c2','c3'],
    "D":['d0','d1','d2','d3'],
}, index=[0,1,2,3])


df2=pd.DataFrame({
    "A":['a4','a5','a6','a7'],
    "B":['b4','b5','b6','b7'],
    "C":['c4','c5','c6','c7'],
    "D":['d4','d5','d6','d7'],
}, index=[4,5,6,7]
)


df3=pd.DataFrame({
    "A":['a8','a9','a10','a11'],
    "B":['b8','b9','b10','b11'],
    "C":['c8','c9','c10','c11'],
    "D":['d8','d9','d10','d11'],
}, index=[8,9,10,11])


# El proposito es concatenar los tres DataFrame
ListaFrame=[df1,df2,df3]
resultado=pd.concat(ListaFrame)
resultado
```

### keys: permite construir un indice jeràquico usando caves que se genera dentro del DataFrae fusionada.

```python
resultado=pd.concat(ListaFrame,keys=["X","Y","Z"])
resultado
```

```python
# visualizamos los datos de la tercera DataFrame
resultado.loc["X"]
```

```python
# Generando DataFrame a partir de los existentes
#LDataFrame=[f*3 for f in ListaFrame]
LDataFrame=list(map(lambda f:f*2,ListaFrame))
resultado=pd.concat(LDataFrame)
resultado
```

```python
LDataFrame=list(map(lambda f: f+"Kg" ,ListaFrame))
resultado=pd.concat(LDataFrame)
resultado
```

```python
## Definamos un cuarto DataFrame
df4=pd.DataFrame({

    "B":['b2','b3','b6','b7'],
    "D":['d2','d3','d6','d7'],
    "F":['f2','f3','f6','f7'],
}, index=[2,3,6,7])
df4
```

```python
# Concatenando en las columnas
resultado=pd.concat([df1,df4],axis=1)
resultado
```

```python
# Concatenando en las filas
resultado=pd.concat([df1,df4],axis=0)
resultado
```

```python
# Aplicando join="inner" se tiene
resultado=pd.concat([df1,df4],axis=1,join="inner")
resultado
```

```python
# Reindexando a partir del DataFrame d1
resultado=pd.concat([df1,df4],axis=1).reindex(df1.index)
resultado
```

```python
# Reindexando a partir del DataFrame d4 previo a la concatenaciòn
resultado=pd.concat([df1,df4.reindex(df1.index)],axis=1).reindex(df1.index)
resultado
```

```python
# Optra forma de concatenar es haciendo uso de append()
resultado=df1.append(df2)
resultado
```

```python
# Concatenando tres dataframe
resultado=df1.append([df2,df3])
resultado
```

```python
# Concatenando df1 y df4 y aplicando ignore_index=True

resultado=pd.concat([df1,df4],ignore_index=True)
resultado
```

```python
# Concatenando df1 y df4 y aplicando ignore_index=True y append()

resultado=df1.append(df4,ignore_index=True)
resultado
```

```python
# Concatenando DataFrame y una serie
sir=pd.Series(["g0","g1","g2","g3"],name="G")
resultado=pd.concat([df1,sir],axis=1)
resultado
```

```python
# Ignorando los nombres de las columnas

resultado=pd.concat([df1,df4],ignore_index=True,axis=1)
resultado
```

```python
# Generado tres Series y luego concatemos y asignamos nombres a las columnas
S1=pd.Series([1,2,3,4])
S2=pd.Series([3,4,5,6])
S3=pd.Series([5,6,6,7])

pd.concat([S1,S2,S3],axis=1,keys=["Edad","Peso","Talla"])

```

```python
# Cancatenando DataFrame almecenados en un diccionario y generando indices para cada uno de ellos
dic={ "X":df1, "Y":df2, "Z":df3}

resultado=pd.concat(dic)
resultado
```

```python
# concatenando algunos dataFrame del diccionario

resultado=pd.concat(dic,keys=["X","Y","Z"])
resultado
```

# Metodo Merge()

Pandas proporciona la funciòn merge(), a fin de realizar todas las operaciones de la union de base de datos entre DataFrame y Series.

```python
izquierda=pd.DataFrame(
    {
        "key":["k0","k1","k2","k3"],
        "A":["a0","a1","a2","a3"],
        "B":["b0","b1","b2","b3"],

    }
)
derecha=pd.DataFrame(
    {
        "key":["k0","k1","k2","k3"],
        "C":["c0","c1","c2","c3"],
        "D":["d0","d1","d2","d3"],

    }
)

resultado=pd.merge(izquierda,derecha, on="key")
resultado
```

```python
# Ejemplo con varias claves de combinanciòn
izquierda=pd.DataFrame(
    {
        "key1":["k0","k0","k1","k2"],
        "key2":["k0","k1","k0","k1"],
        "A":["a0","a1","a2","a3"],
        "B":["b0","b1","b2","b3"],

    }
)
derecha=pd.DataFrame(
    {
        "key1":["k0","k1","k1","k2"],
        "key2":["k0","k0","k0","k0"],
        "C":["c0","c1","c2","c3"],
        "D":["d0","d1","d2","d3"],

    }
)

resultado=pd.merge(izquierda,derecha, on=["key1","key2"])
resultado
```

```python
# Aplicando how="left"
resultado=pd.merge(izquierda,derecha, how="left",on=["key1","key2"])
resultado
```

```python
# Aplicando how="right"
resultado=pd.merge(izquierda,derecha, how="right",on=["key1","key2"])
resultado
```

```python
# Calculando la union en merge()
# Aplicando how="outer"
resultado=pd.merge(izquierda,derecha, how="outer",on=["key1","key2"])
resultado
```

```python

# Aplicando how="inner"
resultado=pd.merge(izquierda,derecha, how="inner",on=["key1","key2"])
resultado
```
