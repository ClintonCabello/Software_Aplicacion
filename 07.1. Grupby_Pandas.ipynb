## Groupby

El m√®todo groupby() permite agrupar filas de datos y llamar a funciones agregadas

```python
import pandas as pd
```

```python
data={ 'Empresa':['Apple','Microsoft','Amazon','Apple','Amazon','Microsoft','Apple'],
        'Nombre': ['Juan','Carla','Karo','Pedro','Luis','Carla','Juan'],
        'Ventas': [2344,3454,2341,4245,4545,5676,6765]}
df=pd.DataFrame(data)
df
```

```python
# se puede usar Grupby para agrupar filas en funcion a nombre de columnas
# por ejemplo vamos a agrupar a partir de la columna empresa
```

```python
df_g=df.groupby('Empresa')
```

# Pivot Table

La principal funci√≤n de la pivot table son las agrupaciones de datos  a las que se le puede aplicar fumciones matem√†ticas.

```python
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
# Haciendo uso de pivot calcular el promedio de personas que sobrevivieron por genero
# Agrupar por genero
# obtener los sobrevivientes
# calcular el promedio
```

```python
titanic.groupby('sex')[['survived']].mean()*100
```

```python
# Promdeio de cuantos sobrevivieron por genero divididos por clase
# agrupar por genero o clase
# obtener los sobrevivientes
# calcular el promedio
# poner el resultado como una tabla
```

```python
titanic.groupby(['sex','class'])['survived'].mean().unstack()
```

```python
# Usando pivot table
```

```python
titanic.pivot_table('survived',index='sex',columns='class')
```

```python
# Los valores marginales
titanic.pivot_table('survived',index='sex',columns='class',margins=True)
```

# Tabla de contingencia

```python
data={ 'Categoria':['principal','asociado','auxiliar','principal','asociado','auxiliar'],
        'Dedicacion': ['DE','TP','DE','TC','TP','TP'],
         'Actividad': ['Investigar','docencia','gestion','Investigar','docencia','gestion'],
          'Nombre': ['Tania','Lili','Luis','Miguel','Jose','Marcia'],
           'TiempoSer': [12,23,14,14,16,24],
           'sueldo': [4000,3000,2500,4000,3000,2500]}
df=pd.DataFrame(data)
df
```

```python
# Generando una tabla de contingencia
pd.crosstab(df['Categoria'],df['Dedicacion'],margins=True)
```

```python
pd.crosstab([df['Categoria'],df['Dedicacion']],df['Actividad'],margins=True)
```
