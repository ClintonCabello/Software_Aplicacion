## Estadística Descriptiva con Python

La estadística es una ciencia que tiene un rol relevante en muchas disciplinas por ello conviene tener nociones básicas al respecto y mejor aún si se domina un software como Python para automatizar análisis estadísticos.

La estadística se clasifica en

        Estadística Descriptiva
        Estadística Inferencial
        
La estadística descriptiva permite obtener estadísticos que representan o describen a los datos. Algunos de los estadísticos más útiles son:
    
    Medidas de tendencia central:
        Media
        Mediana
        Moda
    Medidas de dispersión:
        Desviación media
        Desviación estándar
    Coeficiente de asimetría



El paquete "pandas" contiene las funciones necesarias para realizar análisis descriptivos de los datos.

```python
import numpy as np # importando numpy
import pandas as pd # importando pandas
```

Para realizar los ejemplos de este módulo, se utilizará la tabla "iris", la cual se descarga con la siguiente celda.

```python
# Se importa la base de datos
iris = pd.read_csv("https://raw.githubusercontent.com/toneloy/data/master/iris.csv")
iris.head()
```

```python
print(iris)
#sepal_length: longitud del sépalo
#sepal_width: ancho del sépalo
#petal_length: longitud del pétalo
#petal_width: ancho del pétalo
#species: especie de la flor
```

Se utilizará solo una columna para el análisis

```python
print(iris["petal_width"])
```

```python
# Se puede convertir a lista para observar todos los datos
print(list(iris["petal_width"]))
```

Los estadísticos se pueden obtener a través de métodos:

```python
#Media
iris["petal_width"].mean()
# serie.mean()
```

```python
#Mediana
iris["petal_width"].median()
# serie.median()
```

```python
#Moda
#print(iris["sepal_width"].mode())
print(iris["petal_width"].mode()[0])
# Observe que la moda devuelve una serie de uno o dos elementos y su índice
```

```python
#Varianza
iris["petal_width"].var()
```

```python
#Desviación estandar
iris["petal_width"].std()
```

La asimetría se mide con el coeficiente de Fisher-Person, cuyo resultado tendrá la siguiente interpretación:

     si coeficienteAsimetría = 0: Distribución simétrica
     si coeficienteAsimetría > 0: Distribución asimétrica a la derecha
     si coeficienteAsimetría < 0: Distribución asimétrica a la izquierda

```python
iris["petal_width"].skew()
```

### Gráficos estadísticos

Para los gráficos se utilizará el paquete "seaborn" el cual está basado en "matplotlib" y optimizado para gráficos estadísticos.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Histograma
sns.histplot(iris["petal_width"], kde= True)
#sns.distplot(iris["sepal_width"])
plt.show()
```

```python
# Linea constante
# axvline(constante, estilo de linea, nombre, color, ...)
plt.axvline(iris["petal_width"].mode()[0], linestyle="-", label="mode", color="b")
plt.legend()
```

```python
# Histograma con medidas de tendencia central
plt.figure(figsize=(13,6)) # indicar el tamaño
sns.histplot(iris["petal_width"], kde= True)
#sns.distplot(iris["sepal_width"],color="b",rug=True)
plt.axvline(iris["petal_width"].mean(), linestyle="dashed", label="media", color="k")
plt.axvline(iris["petal_width"].mode()[0],linestyle="dashed", label="moda", color="g")
plt.axvline(iris["petal_width"].median(), linestyle="dashed", label="mediana", color="b")
plt.legend(loc="best",prop={"size":14})
#plt.savefig("ejemploHist.jpg")
plt.show()
```

```python
# Diagrama de la caja
sns.boxplot(x=iris["petal_width"])
```

```python
# Diagrama del violín
sns.violinplot(x = iris["petal_width"])
```

### Correlaciones

La correlación entre dos variables indica la dependencia lineal entre ambas. Con pandas se puede hallar directamente la matriz de correlaciones

```python
iris.corr()
```

Con "pairplot" se genera una matriz con los gráficos de dispersión de todas las variables.

```python
sns.pairplot(iris)
plt.show()
```

A la par también se pueden hallar correlaciones y diagramas de dispersión para cada par de variables por separado.

```python
#Obtener una correlación específica entre dos varialbes
iris.corr()["petal_length"]["sepal_length"]
```

```python
# Graficar el diagrama de dispersión entre dos variables
fig, ax = plt.subplots()
ax.scatter(iris["petal_length"], iris["petal_length"])
plt.show()
# Pueden revisar el blog si desea personalizar aún más el gráfico de dispersión
# https://www.interactivechaos.com/manual/tutorial-de-matplotlib/graficos-de-dispersion
```

Otro gráfico interesante es el que se obtiene al agregar una clasificación de datos en el método "pairplot".

```python
# Graficar los diagramas de dispersión añadiendo la clasificación de datos
sns.pairplot(iris, hue="species")
plt.show()
# el argumento "hue" divide los datos según la fila que se indique y les asigna diferentes colores
# https://seaborn.pydata.org/generated/seaborn.pairplot.html
```

La celda anterior permite observar la correlación de manera visual, mientras que la celda posterior permite obtener el valor numérico de la correlación.

```python
#Mapa de color de las correlaciones:

#se obtienen las correlaciones
correlaciones = iris.corr()

#se crea una máscara
mask = np.zeros_like(correlaciones) #indicar los datos
mask[np.triu_indices_from(mask)] = True

plt.figure(figsize=(7,5))

with sns.axes_style("white"):
    sns.heatmap(correlaciones,annot=True,linewidth=5,
                mask = mask,cmap="summer")
plt.title("Correlación entre variables")
plt.show()
```

3. Cree una función que tenga como entrada una serie cualquiera de pandas (es decir una columna de datos) y que tenga como salida las medidas de tendencia central y de desviación.

```python
def estadistica(serie):
  med=serie.mean()
  mediana=serie.median()
  vari=serie.var()
  stan=serie.std()
  return [med,mediana,vari,stan]
print(estadistica(iris['petal_length']))
```

Cree una función que tenga como entrada una serie cualquiera de pandas (es decir una columna de datos) y que tenga como salida el histograma, el gráfico de la caja y el gráfico del violín.

```python
def graficos(serie):
  #Histograma

  fig, ((ax1), (ax2),(ax3)) = plt.subplots(3, 1)

  ax1.hist(serie)
  ax2.boxplot(serie)
  ax3.violinplot(serie)
  plt.show()
graficos(iris['petal_width'])
```

Cree una función que tenga como entrada un dataframe de pandas y devuelva: las medidas de tendencia central, las de dispersión, el histograma, el diagrama de la caja y el diagrama del violín de cada una de las variables. Tenga en cuenta que se tomará por hecho que las variables que se pasan como argumento son variables cuantitativas en formato numérico (enteros o flotantes)

```python
def descriptiva(df):
  des=df.describe()
  print(des)
  sns.pairplot(df)
  for i in range(4):
    fig, ((ax1), (ax2),(ax3)) = plt.subplots(3, 1)

    ax1.hist(df.loc[i])
    ax2.boxplot(df.loc[i])
    ax3.violinplot(df.loc[i])
  plt.show()

descriptiva(iris[["sepal_length",  "sepal_width",  "petal_length", "petal_width"]])
```
